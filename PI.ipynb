```python
import csv
import pandas as pd
```

```python
data = open ('Flight Distance Test.csv',encoding = 'utf-8')

csv_data = csv.reader(data)

Doc= list (csv_data)

# 1) opens the file 
# 2) translates the csv
# 3) translates csv into list of lists
```

```python
test_df = pd.read_csv('Flight Distance Test.csv')
test_df.head()
```

```python
# Create empty lists as columns to add to dataframe later

Normalised_City_Pair = []
Departure_Code = []
Arrival_Code = []
Departure_lat = []
Departure_lon = []
Arrival_lat = []
Arrival_lon = []
```

```python
# Create a function to make the numbers type float and append to lists

def add_to_list(list_name,index):
    for item in Doc[1:]:
        list_name.append(item[index])
        
def add_to_list2(list_name,index):
    for item in Doc[1:]:
        list_name.append(float(item[index]))
```

```python
add_to_list(Normalised_City_Pair,0)
add_to_list(Departure_Code,1)
add_to_list(Arrival_Code,2)
add_to_list2(Departure_lat,3)
add_to_list2(Departure_lon,4)
add_to_list2(Arrival_lat,5)
add_to_list2(Arrival_lon,6)
```

```python
d ={'Normalised_City_Pair':Normalised_City_Pair,'Departure_Code':Departure_Code,'Arrival_Code':Arrival_Code,'Departure_lat':Departure_lat,'Departure_lon':Departure_lon,'Arrival_lat':Arrival_lat,'Arrival_lon':Arrival_lon}
```

```python
df = pd.DataFrame (d, columns = ['Normalised_City_Pair','Departure_Code','Arrival_Code','Departure_lat','Departure_lon','Arrival_lat','Arrival_lon'])
# delete a single row by index value 0
#df = df.drop(labels=0, axis=0)
#df.head()
```

```python
# Finding the distance between airports using Pythagoras' theorem
df['Distance'] = ((df['Arrival_lat'] - df['Departure_lat'])**2 + (df['Arrival_lon'] - df['Departure_lon'])**2)**0.5
df.head()
```
