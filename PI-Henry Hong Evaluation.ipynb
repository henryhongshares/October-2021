```python
import csv
import pandas as pd
import numpy as np
```

```python
# Open the file as a dataframe and then change the column names so we have no issues referencing them later

test_df = pd.read_csv('Flight Distance Test.csv')
test_df= test_df.rename(columns={"Departure Code": "Departure_Code", "Arrival Code": "Arrival_Code"})
test_df.head()
```

```python
# Finding the distance between airports using the Pythagorean theorem

test_df['Distance'] = ((test_df['Arrival_lat'] - test_df['Departure_lat'])**2 + (test_df['Arrival_lon'] - test_df['Departure_lon'])**2)**0.5
test_df.head()
```

```python
# Top 10 Arrival destinations

popular_destinations_df = pd.DataFrame(test_df.Arrival_Code.value_counts().reset_index().values, columns=["Arrival", "Total"])
popular_destinations_df = popular_destinations_df.sort_index(axis = 0, ascending=True)
popular_destinations_df[0:10].plot('Arrival','Total','bar')
popular_destinations_df[0:10]
```

```python
# Replace all LHR departure latitude values so they now equal 51.5 and longitude -0.45

new_list=[]
for item in test_df['Departure_lat']:
    new_list.append(51.5)
    
new_list2=[]
for item in test_df['Departure_lon']:
    new_list2.append(-0.45)
      
test_df['Departure_lat']=new_list
test_df['Departure_lon']=new_list2


# Now re-compute the distance between departure and arrival

test_df['Distance'] = ((test_df['Arrival_lat'] - test_df['Departure_lat'])**2 + (test_df['Arrival_lon'] - test_df['Departure_lon'])**2)**0.5
test_df[0:15]
```
