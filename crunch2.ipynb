```python
def convert_to_float_BTC(x,y,z):
    for row in x[-lookback_period+2:]:
        z.append(float(row[y]))

def convert_to_float_OTHERS(x,y,z):
    for row in x[-lookback_period:-2]:
        z.append(float(row[y]))
BTC_open_price= []
BTC_close_price= []
BTC_high_price= []
BTC_low_price= []
ETH_high_price= []
ETH_low_price= []
DOT_high_price= []
DOT_low_price= []
LTC_high_price= []
LTC_low_price= []
LINK_high_price= []
LINK_low_price= []
AAVE_high_price= []
AAVE_low_price= []
XRP_high_price= []
XRP_low_price= []

convert_to_float_BTC(BTC [1:],1,BTC_open_price)
convert_to_float_BTC(BTC [1:],4,BTC_close_price)
convert_to_float_BTC(BTC [1:],2,BTC_high_price)
convert_to_float_BTC(BTC [1:],3,BTC_low_price)
convert_to_float_OTHERS(ETH [1:-2],2,ETH_high_price)
convert_to_float_OTHERS(ETH [1:-2],3,ETH_low_price)
convert_to_float_OTHERS(DOT [1:-2],2,DOT_high_price)
convert_to_float_OTHERS(DOT [1:-2],3,DOT_low_price)
convert_to_float_OTHERS(LTC [1:-2],2,LTC_high_price)
convert_to_float_OTHERS(LTC [1:-2],3,LTC_low_price)
convert_to_float_OTHERS(LINK [1:-2],2,LINK_high_price)
convert_to_float_OTHERS(LINK [1:-2],3,LINK_low_price)
convert_to_float_OTHERS(AAVE [1:-2],2,AAVE_high_price)
convert_to_float_OTHERS(AAVE [1:-2],3,AAVE_low_price)
convert_to_float_OTHERS(XRP [1:-2],2,XRP_high_price)
convert_to_float_OTHERS(XRP [1:-2],3,XRP_low_price)
```

```python
lows_df = pd.DataFrame(BTC_date_list,columns =["BTC_date_list"])
lows_df ["BTC_low_price"] = BTC_low_price
lows_df ["BTC_rolling_low"] = lows_df['BTC_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["ETH_low_price"] = ETH_low_price
lows_df ["ETH_rolling_low"] = lows_df['ETH_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["DOT_low_price"] = DOT_low_price
lows_df ["DOT_rolling_low"] = lows_df['DOT_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["LTC_low_price"] = LTC_low_price
lows_df ["LTC_rolling_low"] = lows_df['LTC_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["LINK_low_price"] = LINK_low_price
lows_df ["LINK_rolling_low"] = lows_df['LINK_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["AAVE_low_price"] = AAVE_low_price
lows_df ["AAVE_rolling_low"] = lows_df['AAVE_low_price'].rolling(rolling_period).min().shift(+1)
lows_df ["XRP_low_price"] = XRP_low_price
lows_df ["XRP_rolling_low"] = lows_df['XRP_low_price'].rolling(rolling_period).min().shift(+1)

lows_df ["BTC_LOWER_low"] = np.where(lows_df['BTC_low_price'] < lows_df['BTC_rolling_low'],1,0)
lows_df ["ETH_LOWER_low"] = np.where(lows_df['ETH_low_price'] > lows_df['ETH_rolling_low'],1,0) 
lows_df ["DOT_LOWER_low"] = np.where(lows_df['DOT_low_price'] > lows_df['DOT_rolling_low'],1,0) 
lows_df ["LTC_LOWER_low"] = np.where(lows_df['LTC_low_price'] > lows_df['LTC_rolling_low'],1,0) 
lows_df ["LINK_LOWER_low"] = np.where(lows_df['LINK_low_price'] > lows_df['LINK_rolling_low'],1,0) 
lows_df ["AAVE_LOWER_low"] = np.where(lows_df['AAVE_low_price'] > lows_df['AAVE_rolling_low'],1,0) 
lows_df ["XRP_LOWER_low"] = np.where(lows_df['XRP_low_price'] > lows_df['XRP_rolling_low'],1,0) 

column_list =['BTC_LOWER_low','ETH_LOWER_low','DOT_LOWER_low','LTC_LOWER_low','LINK_LOWER_low','AAVE_LOWER_low','XRP_LOWER_low']
lows_df ["total_lower_lows"] = lows_df [column_list].sum(axis=1)
lows_df
```
