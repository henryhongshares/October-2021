```python
rolling_period_A = 240
rolling_period_B = 60
rolling_period_C = 20
```

```python
#DROPS IRRELEVANT COLLUMNS
BTC = BTC.drop(BTC.columns[[1,4,5,6,7,8,9,10,11]], axis=1)  
ETH = ETH.drop(ETH.columns[[1,4,5,6,7,8,9,10,11]], axis=1)  
DOT = DOT.drop(DOT.columns[[1,4,5,6,7,8,9,10,11]], axis=1)  
LTC = LTC.drop(LTC.columns[[1,4,5,6,7,8,9,10,11]], axis=1) 
LINK = LINK.drop(LINK.columns[[1,4,5,6,7,8,9,10,11]], axis=1) 
AAVE = AAVE.drop(AAVE.columns[[1,4,5,6,7,8,9,10,11]], axis=1)  
XRP = XRP.drop(XRP.columns[[1,4,5,6,7,8,9,10,11]], axis=1) 
```
